###Â 2. Analyse and plot the data

```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

**Box Plot**

```python
ax = sns.boxplot(data = brics, x = 'expectancy', y = 'population')
```

**Scatter Plot**

```python
plt.figure(figsize=(8, 6))
plt.scatter(brics['population'], brics['expectancy'], color='royalblue')
plt.xlabel('Population (in millions)')
plt.ylabel('Life Expectancy (years)')
plt.title('Life Expectancy vs Population - BRICS')
plt.grid(True)

for i, row in brics.iterrows():
    plt.text(row['population'], row['expectancy'], row['country'], fontsize=9)

plt.show()
```

**Bubble Plot**

```python
plt.scatter(brics['population'], brics['expectancy'], s=brics['gdp'], alpha=0.5)
```

**Regression Plot**

```python
sns.regplot(x='population', y='expectancy', data=brics)
```
